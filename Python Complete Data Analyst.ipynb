### Conditional Statements (if, elif, else)
- if :- The if statement checks if a condition is True. If it is, the block of code indented under the if statement is executed.
- else :- The else statement runs a block of code if none of the previous conditions are True.
- elif :- The elif (short for "else if") statement is used to check another condition if the previous if condition is False.

```python
# if condition
age = 25
if age >= 18:
    print("You are eligible to vote")
```

```python
# else statement :- The else statement  block of code ig the condition in the if statement is false
age = 14
if age >= 18:
    print("You are eligible to vote")
else:
    print("You can't vote")
```

```python
# elif :- The elif statement allows you to check multiple conditions. It stands for "else if"
age = int(input("Enter your age :"))
if age < 13:
    print("You are a child")
elif age < 18:
    print("You are a minor")
else:
    print("You are an adult and you can take decision by youself")
```

Nested conditional statement :- You can place one or more if, elif, or else statements inside another if, elif, or else statement to create nested conditional statements.

```python
# Number is even, odd, negative
num = int(input("Enter a number : "))
if num > 0:
    print("The number is positive")
    if num % 2 == 0:
        print("Your number is even")
    else:
        print("Your number is odd")
else:
    print("Number is zero or negative")
```

Loop Control Statements :-
1. break :- The break statement exits the loop prematurely
2. continue :- The continue statement skips the current iteration and continues with the next.
3. pass :- The pass statement is a null operation; it does nothing.

```python
# break statement
for i in range(10):
    if i == 5:
        break
    print(i)
```

```python
# Continue statement
for i in range(10):
    if i%2 == 0:
        continue
    print(i)
```

```python
# Pass statement
for i in range(10):
    if i == 3:
        print("the number is",i)
        pass
    print(i)
```

- Nested Loops :- Loop inside a loop

```python
for i in range(3):
    for j in range(2):
        print(f"i:{i} and j:{j}")
```

##### List Comprehension
- Basics Syantax        :    [expression for item in iterable]
- with conditional logic  :  [expression for item in iterable if condition]
- Nested List Comprehension  : [expression for item1 in iterable1 for item2 in iterable2]


```python
# Basic List Comprehension
square = [num ** 2 for num in range(10)]
print(square)
```

```python
# List Comprehension with Condition
lst=[]
for i in range(10):
    if i % 2 == 0:
        lst.append(i)
print(lst)
```

```python
even_numbers=[num for num in range(10) if num%2==0]
print(even_numbers)
```

```python
# Nested List Comprehension
lst1 = [1,2,3,4]
lst2 = ["a", "b", "c", "d"]
pair = [[i,j] for i in lst1 for j in lst2]
print(pair)
```

```python
# List comprehension with function calls
words = ["hello", "world", "python", "list", "comprehension"]
lengths = [len(word) for word in words]
print(lengths)
```

2. Password Strength Checker

```python
def is_strong_password(password):
    """This function checks if the password is strong or not"""
    if len(password)<8:
        return False
    if not any(char.isdigit() for char in password):
        return False
    if not any(char.islower() for char in password):
        return False
    if not any(char.isupper() for char in password):
        return False
    if not any(char in '!@#$%^&*()_+' for char in password):
        return False
    return True

## calling the function
print(is_strong_password("WeakPwd"))
print(is_strong_password("Str0ngPwd!"))
print(is_strong_password("jhgfdfghj"))    
```

#### Interitance :-
- Inheritance in Python is a mechanism that allows a class (child class) to acquire properties and methods from another class (parent class). It promotes code reuse and helps implement polymorphism and abstraction in object-oriented programming.

- When you class(child/derived) derives the properties and methods of another class (parent/class)
     - class Car:
          - ....
     - class ToyotaCar(Car):
          - ....

```python
class Car:
    @staticmethod
    def start():
        print("car started...")
        
    @staticmethod
    def stop():
        print("car stopped...")

class ToyotaCar(Car):
    def __init__(self, name):
        self.name = name

car1 = ToyotaCar("Fortuner")
car2 = ToyotaCar("Land Cruisor")

print(car1.name)
print(car1.start())
print(car2.name)
print(car2.start())
```

#### Types of Inheritance :-
- Single Inheritance :- A child class inherits from a single parent class.
- Multi-level Inheritance :- A class inherits from another child class.
- Multiple Inheritance :- A child class inherits from multiple parent classes.
- Hierarchical Inheritance: Multiple child classes inherit from one parent class.
- Hybrid Inheritance: A mix of two or more types of inheritance.

```python
class A:
    varA = "welcome to class A"

class B:
    varB = "welcome to class B"

class C(A, B):
    varC = "welcome to class C"

c1 = C()

print(c1.varC)
print(c1.varB)
print(c1.varA)
```

- Access Parent Class Methods: It allows calling methods or constructors of the parent class without directly referencing the class name.
- Avoids Hardcoding: Works dynamically with parent class methods, making it easier to work with multiple inheritance and maintain code.
- Single or Multiple Inheritance: It's primarily used to resolve the Method Resolution Order (MRO) in multiple inheritance.

```python
class Parent:
    def __init__(self, name):
        self.name = name

    def greet(self):
        print(f"Hello, {self.name}!")


class Child(Parent):
    def __init__(self, name, age):
        # Use super() to call the parent class constructor
        super().__init__(name)
        self.age = age

    def greet(self):
        super().greet()  # Call Parent's greet method
        print(f"I am {self.age} years old!")
```

```python
class Vehicle:
    def __init__(self, brand):
        self.brand = brand

    def start(self):
        print(f"{self.brand} vehicle is starting!")

class Car(Vehicle):
    def __init__(self, brand, model):
        super().__init__(brand)  # Call the Vehicle constructor
        self.model = model

    def start(self):
        super().start()  # Call the Vehicle start method
        print(f"The model is {self.model}")

# Using the classes
my_car = Car("Toyota", "Corolla")
my_car.start()
```

### Property decorator :-
- We use @property decorator on any method in the class to use the method as a property.
- The @property decorator in Python is a way to define a method that acts like an attribute. It allows you to encapsulate instance attributes, providing controlled access to them. With @property, you can define getter, setter, and deleter methods while using them like simple attributes, making the code cleaner and more intuitive.

##### Key Feature :-
- Encapsulation :- Control access to an attribute by adding logic to get, set, or delete it.
- Attribute-Like Access :- Methods decorated with @property are accessed like attributes without parentheses.
- Read-Only or Writable :- Define only a getter for read-only properties or both getter and setter for writable properties.

```python
class Student:
    def __init__(self, phy, chem, math):
        self.phy = phy
        self.chem = chem
        self.math = math
       
    @property
    def percentage(self):
        return str((self.phy + self.chem + self.math) / 3) + "%"
        
   # def calcPercentage(self):
    #    self.percentage = str((self.phy + self.chem + self.math) / 3) + "%"

stu1 = Student(98, 97, 99)
print(stu1.percentage)

stu1.phy = 86
print(stu1.percentage)
```

### Polymorphism : Operator Overloading
- When the same operator is allowed to have different meaning according to the context.
- Polymorphism in Python allows objects of different types to be treated uniformly. One way polymorphism is implemented in Python is through operator overloading, which allows us to define custom behavior for standard operators (like +, -, *, etc.) when applied to user-defined objects.



```python
# Operator and Dunder funciton :-
   # - a+b # addition          - a.__add__(b)
   # - a-b # substraction      - a.__sub__(b)
   # - a*b # multiplication    - a.__mul__(b)
   # - a/b #division           - a.__truediv__(b)
   # - a%b #modulus            - a.__mod___(b)
```

#### What is Operator Overloading?
By default, operators like +, -, *, etc., work with built-in data types (e.g., integers, floats). With operator overloading, we can redefine these operators for custom objects by implementing specific special methods (dunder methods) in a class.

```python
class Vector:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        return Vector(self.x + other.x, self.y + other.y)

    def __repr__(self):
        return f"Vector({self.x}, {self.y})"

# Usage
v1 = Vector(1, 2)
v2 = Vector(3, 4)
v3 = v1 + v2  # Calls __add__()
print(v3)  
```

```python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __mul__(self, scalar):
        return Point(self.x * scalar, self.y * scalar)

    def __repr__(self):
        return f"Point({self.x}, {self.y})"

# Usage
p = Point(2, 3)
scaled = p * 4  # Calls __mul__()
print(scaled)  
```

```python
class Box:
    def __init__(self, volume):
        self.volume = volume

    def __lt__(self, other):
        return self.volume < other.volume

    def __eq__(self, other):
        return self.volume == other.volume

# Usage
box1 = Box(10)
box2 = Box(20)
print(box1 < box2)  # Output: True
print(box1 == box2) # Output: False
```

```python
class ComplexNumber:
    def __init__(self, real, imag):
        self.real = real
        self.imag = imag

    def __add__(self, other):
        return ComplexNumber(self.real + other.real, self.imag + other.imag)

    def __mul__(self, other):
        real = self.real * other.real - self.imag * other.imag
        imag = self.imag * other.real + self.real * other.imag
        return ComplexNumber(real, imag)

    def __repr__(self):
        return f"{self.real} + {self.imag}i"

# Usage
c1 = ComplexNumber(1, 2)
c2 = ComplexNumber(3, 4)
print(c1 + c2)  
print(c1 * c2)  
```

```python
#Read a text file and count the number of lines, words, and characters.
# Counting lines, words, and characters in a text file
def count_text_file(file_path):
    with open(file_path, 'r') as file:
        lines = file.readlines()
        line_count = len(lines)
        word_count = sum(len(line.split()) for line in lines)
        char_count = sum(len(line) for line in lines)
    return line_count, word_count, char_count

file_path = 'examples.txt'
lines, words, characters = count_text_file(file_path)
print(f'Lines: {lines}, Words: {words}, Characters: {characters}')
```

```python
data_df = pd.read_csv('data.csv')
print(data_df.head(5))
```

```python
# Print sales information
print(data_df['Sales'])
```

```python
# Print summary statistics for Sales
print(data_df['Sales'].describe())
```

```python
# Line plot of Sales over Date
data_df.plot(x='Date', y='Sales', kind='line', title='Sales Over Time')
plt.show()
```

```python
# Histogram of Sales
data_df['Sales'].plot(kind='hist', bins=10, title='Sales Distribution')
plt.show()
```

```python
# Area plot of Sales
data_df.plot(x='Date', y='Sales', kind='area', title='Area Plot of Sales Over Time')
plt.show()
```

```python
# Hexbin plot of Value vs Sales
data_df.plot.hexbin(x='Value', y='Sales', gridsize=25, title='Hexbin Plot of Value vs Sales')
plt.show()
```

```python
# Stacked bar plot for Sales by Region and Product
data_df.groupby(['Sales', 'Region', 'Product'])['Sales'].sum().unstack().plot(kind='bar', stacked=True, title='Stacked Bar Plot of Sales by Region and Product')
plt.show()
```

```python
import seaborn as sns
```

```python
### Basic Plotting With Seaborn
tips=sns.load_dataset('tips')
tips
```

```python
# Create a scatter plot
import matplotlib.pyplot as plt
sns.scatterplot(x = 'total_bill', y = 'tip', data=tips)
plt.title("Scatter Plot of Total Bill vs Tip")
plt.show()
```

```python
# Line Plot
sns.lineplot(x='size', y='total_bill', data = tips)
plt.title("Line Plot of total bill size")
plt.show()
```

```python
# Categorical Plot - Bar Plot
sns.barplot(x = 'day', y = 'total_bill', data = tips)
plt.title('Bar plot of total bill')
plt.shoW()
```

```python
## Box Plot
sns.boxplot(x="day",y='total_bill',data=tips)
```

```python
##  ViolinPlot
sns.violinplot(x="day",y='total_bill',data=tips)
```

```python
### Histograms
sns.histplot(tips['total_bill'],bins=10,kde=True)
```

```python
## KDE Plot
sns.kdeplot(tips['total_bill'],fill=True)
```

```python
# Pairplot
sns.pairplot(tips)
```
